```python
import tensorflow as tf

from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
import keras

import pandas as pd
import numpy as np
from PIL import Image 
import cv2

from joblib import dump, load
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure

import matplotlib.pyplot as plt
```

```python
def output(n):
    emotion= {0:"angry",1:"disgust",2:"fear",3:"happy",4:"neutral",5:"sad",6:"surprise"}
    print(emotion[n[0]])
    
```

HÌNH TRONG TẬP TEST

```python
test_image = image.load_img('E:/Code files/ipynb file/test/happy/PrivateTest_95094.jpg', 
                            target_size = (128, 128))

test_image = image.img_to_array(test_image)

plt.imshow(test_image/255)
test_image = np.expand_dims(test_image, axis = 0)
```

```python
test_image = image.load_img('E:/Code files/ipynb file/test/happy/PrivateTest_95094.jpg', 
								target_size = (48, 48), color_mode="grayscale")
test_image = image.img_to_array(test_image)
test_image = test_image / 255
test_image, hog_image = hog(test_image, orientations=9, pixels_per_cell=(8, 8),
                	cells_per_block=(2, 2), visualize=True, multichannel=True)

test_image = np.array(test_image)
np.expand_dims(test_image, axis = 0)
test_image = test_image.reshape(1,-1)
```

```python
clf_rbf = load('model_svm_clf.joblib')
clf_pl = load('model_svm_clf_pl.joblib')
output(clf_rbf.predict(test_image))
output(clf_pl.predict(test_image))
```
